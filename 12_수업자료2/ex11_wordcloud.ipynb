# WordCloud

* 메타 데이터에서 얻어진 태그들을 분석하여 중요도나 인기도 등을 고려하여 시각적으로 늘어놓아 표시하는 것
* 시각적인 중요도를 강조를 위해 각 태그들은 그 중요도(혹은 인기도)에 따라 글자의 색상이나 굵기 등 형태가 변한다.

```python
import matplotlib.pyplot  as plt
from wordcloud import WordCloud, STOPWORDS
import numpy as np
from PIL import Image

text="""
Life is just a bowl of cherry
Sometimes it is sweet or filled with worries
Don't be afraid
When things go wrong, just be strong
When things seems up in the air
And everything is so unfair
And you stumble and fall
Just pick yourself up and sing
If one day you lose your way
Just remember one thing, my friend
When you're under a cloud
Just visit music and sing
If one day you lose your way
Just remember that I'm here to stay
Don't you give up
Keep your chin up and be happy
Life is just a bowl of cherry
Sometimes it is sweet or filled with worries
Don't be afraid
When things go wrong, just be strong
When things seems up in the air
And everything is so unfair
And you stumble and fall
Just pick yourself up and sing
If one day you lose your way
Just remember one thing, my friend
When you're under a cloud
Just visit music and sing
If one day you lose your way
Just remember one thing, my friend
When you're under a cloud
Just visit music and dance
If one day you lose your way
Just remember that I'm here to stay
Don't you give up, keep your chin up
Don't you give up, keep your chin up
Keep your chin up and be happy

"""
spwords=set(STOPWORDS) # 기본적으로 많이 쓰는 단어 (제외할 단어)
                       # ex)'should', "wouldn't", "how's", "i'm", 'then', etc ..
#spwords.add('up')    #제외하고 싶은 단어 추가
spwords.discard('up')
print(spwords)
```

```python
mask = np.array(Image.open('./data/mask.png'))

wordcloud = WordCloud(max_font_size=200,
                      stopwords=spwords,   #기본적으로 많이 쓰는 단어는 제외
                      background_color='#FFFFFF',
                      #random_state=50,    #워드클라우드 모양 결정
                      #colormap="plasma",  # 
                      mask=mask,
                      font_path = 'C:\Windows\Fonts\malgunbd.ttf',
                      width=1000,
                      height=800).generate(text)
```

```python
plt.figure(figsize=(10,8))
plt.imshow(wordcloud)
plt.axis('off')

plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 

from wordcloud import WordCloud, STOPWORDS

import numpy as np
from PIL import Image ##이미지 관련 함수

text=open("data/remember.txt","r", encoding="utf-8").read()

spwords=set(STOPWORDS) ## 기본적으로 많이 쓰는 단어는 제외
spwords.add('10월') ##  제외하고 싶은 단어 추가


alice_mask=np.array(Image.open('data/mask.png'))
```

```python
# 워드 클라우드를 설정합니다. 한글이 있으면 폰트도 넣어 줍니다.
wordcloud = WordCloud(max_font_size=200,
                      #font_path='c:/Windows/fonts/malgun.ttf',
                      stopwords=spwords,
                      background_color='#FFFFFF',
                      max_words = 2000,  
                      mask=alice_mask,
                      width=1200,
                      height=800).generate(text)

plt.figure(figsize=(10,8))
plt.imshow(wordcloud)
plt.tight_layout(pad=0)
plt.axis('off')

plt.savefig('wordcloud_image3.png', bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 

from wordcloud import WordCloud, STOPWORDS
text=open("data/remember.txt","r", encoding="utf-8").read()

spwords=set(STOPWORDS) ## 기본적으로 많이 쓰는 단어
spwords.add('10월') ##  제외하고 싶은 단어 추가

wordcloud = WordCloud(max_font_size=200,
                      #font_path='c:/Windows/fonts/malgun.ttf',
                      stopwords=spwords,
                      background_color='#FFFFFF',
                      max_words = 2000,
                      colormap="plasma",
                      width=1200,
                      height=800).generate(text)
```

```python
plt.rcParams["font.family"] = 'DejaVu Sans'
plt.rcParams["font.size"] = 20
plt.rcParams["figure.figsize"] = (14,4)

plt.figure(figsize=(10,8))
plt.imshow(wordcloud)
plt.tight_layout(pad=0)
plt.axis('off')


plt.savefig('wordcloud_image4.png', bbox_inches='tight')
plt.show()
```
