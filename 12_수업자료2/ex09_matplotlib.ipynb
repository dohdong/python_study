## 1. 소개

### (1) 구조

* Figure : 그래프 전체 그림
* Axes : 그래프가 그려지는 좌표 평면. Figure 의 subplot 이 Axes임. 한개의 Figure 에 여러개의 Axes 가 올 수 있음
* X axis : X 축
* Y axis : Y 축
* Tick : Major 눈금과 Minor 눈금이 있음
* Grid : 그래프 격자
* Title : 그래프 제목
* Label : 각 축이나 눈금 등에 붙이는 이름
* Legend : 각 그래프의 이름(label)을 알기 쉽게 별도로 표시해 놓는 것

### (2) 그래프 종류

|함수|설명|
----|-----
bar|	막대 플롯(bar plot) 생성
hist|	히스토그램 플롯
plot|	플롯 라인 또는 마커
scatter|	x대 y의 산점도 생성
errorbar|	오류 막대 그래프
boxplot|	박스플롯 생성
violinplot|	바이올린 플롯
xcorr|	cross corr
acorr|	auto corr

## 5. Subplots

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize = (10,10))
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 2, 4, 10])

plt.subplot(2, 2, 1) # 1행 2열 중에 첫번째
plt.plot(x,y,'r-')
plt.title('gg')
#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 3, 40])

plt.subplot(2, 2, 2) 
plt.plot(x,y)
plt.title('hh')

x = np.array([0, 1, 2, 3])
y = np.array([1, 2, 3, 40])

plt.subplot(2, 2, 3) 
plt.plot(x,y)
plt.title('kk')


plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(19680801)    # Reproducible random state
# np.random.seed()를 이용해서 난수 생성의 시드를 설정하면, 같은 난수를 재사용할 수 있습니다.
# seed()에 들어갈 파라미터는 0에서 4294967295 사이의 정수여야 합니다.

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()


# scatter()에 x, y 위치를 입력해줍니다.
# s는 마커의 면적을, c는 마커의 색을 지정합니다.
# alpha는 마커 색의 투명도를 결정합니다.
```

```python
import matplotlib.pyplot as plt
from csv import reader

with open("data/pop_Seoul.csv", 'r', encoding='utf-8') as f :
       
    years = [] # 년도 리스트 
    populations = [] # 인구수 리스트

    header = next(f)
    
    for line in f : 
        (year, population) = line.split(',')
        years.append(year)
        populations.append(int(population))

    print(years)
    print(populations)
```

```python
# 그래프를 그릴 도화지 준비하기
fig = plt.figure(figsize=(10, 8))

plt.plot(df['year'].astype(str), df['population'], color='red', marker='o', linestyle='dashed')

plt.title("Seoul Population Change")

plt.xlabel('year')
plt.ylabel("10Million")
plt.show()
```

```python
fig = plt.figure(figsize=(10, 10))

plt.plot(df.loc[:, 'year'], df.loc[:,'population'], color='green', marker='o', linestyle='solid')

plt.title("Seoul Population Change")

plt.xticks(df.loc[:, 'year'], labels=df.loc[:, 'year'])
plt.ylabel("10Million")
plt.show()
```

```python
fig = plt.figure()

plt.bar(df['year'].astype(str), df['population'])

plt.title("Seoul Population Change")

plt.ylabel("10Million")
plt.show()
```

```python
fig = plt.figure(figsize=(10, 10))

plt.bar(df.loc[:, 'year'], df.loc[:,'population'], width=3)

plt.title("Seoul Population Change")

plt.xticks(df.loc[:, 'year'], labels=df.loc[:, 'year'])
plt.ylabel("10Million")
plt.show()
```
