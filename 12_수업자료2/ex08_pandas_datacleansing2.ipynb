# 2. 이상치

```python
import pandas as pd
 
df = pd.read_csv('data/dirtydata.csv', encoding="utf-8") # 26 행 Date 형식 / 7 행 Duration 잘못된 데이터 / 11, 12 행 중복

df
```

```python
import pandas as pd

df = pd.read_csv('data/dirtydata.csv')

df['Date'] = pd.to_datetime(df['Date']) # 26 행 날짜 형식 변경

df
```

```python
df.dropna(subset=['Date'], inplace = True) ## 22번 행 제거

df
```

```python
df.loc[7, 'Duration'] = 45 # 450 -> 45
df
```

```python
for x in df.index:
    if df.loc[x, "Duration"] > 120:
        df.loc[x, "Duration"] = 120 # 잘못된 값 변경
        
df
```

```python
for x in df.index:
    if df.loc[x, "Duration"] > 120:
        df.drop(x, inplace = True)  # 잘못된 값 변경
df
```

```python
df.duplicated() # 중복 여부 확인
```

```python
df.drop_duplicates(inplace = True) # 중복 제거

df
```

```python
df['MaxP'] = pd.cut(x = df['Maxpulse'], bins=[100, 120, 150], labels=['Low', 'High'])
df
```
