## 3. 사고, 날씨 데이터 로드

```python
rA = pd.read_csv('realAll.csv') # 출력하면 아래와 같다.
rA.head()
```

### 3.1 가공

```python
rA = rA.replace('(.*)\n(.*)', r'\1 \2', regex=True)
rA = rA.replace('(.*)눈(.*)', r'\1 1\2', regex=True)
rA = rA.replace('(.*)비(.*)', r'\1 2\2', regex=True)
rA = rA.replace('[^0-9]', '', regex=True)
rA = rA.replace('(.*)12(.*)', r'\1 3\2', regex=True)
rA = rA.replace('(.*)21(.*)', r'\1 3\2', regex=True) 
    # 마찬가지로 가공을 통해 원하는 모습으로 만들어주었다
```

## 4. 결과 막대 그래프

```python
import matplotlib.pyplot as plt # 2019년의 다양한 날씨에 따른 사고율 그래프이다.
import numpy as np

plt.rc('font', family='Malgun Gothic')

fig = plt.figure(figsize=(10, 6))


x = np.array(['눈,비','구름만 낀날','날씨가 있는 날','비', '눈', '날씨가 없는 날'])
y = np.array([116.12,105.10,106.65,107.89,106.54,109.19])

plt.ylim([80,120])

plt.bar(x,y)
plt.show()
```

```python
plt.rc('font', family='Malgun Gothic') # 2019년의 그래프이다
fig = plt.figure(figsize=(5, 6))

x = np.array(['날씨가 있는 날', '날씨가 없는 날'])
y = np.array([106.64,109.25])

plt.ylim([60,120])
plt.bar(x,y)
plt.show()
```

```python
#plot 1:               # 2019년과 2020년의 비교 그래프이다.
plt.rc('font', family='Malgun Gothic')
fig = plt.figure(figsize=(10,6))
x = np.array(['2019', '2020'])
y = np.array([106.64,94.94])
plt.subplot(1, 2, 1)
plt.ylim([90,110])
plt.title('날씨가 있는 날')
plt.bar(x,y,width = 0.7,color = 'dodgerblue')

#plot 2:
x = np.array(['2019', '2020'])
y = np.array([109.25,99.37])
plt.subplot(1, 2, 2) 
plt.ylim([90,110])
plt.title('날씨가 없는 날')
plt.bar(x,y,width = 0.7,color = 'springgreen')

plt.show()
```
