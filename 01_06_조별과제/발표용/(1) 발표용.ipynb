```python
#plot 1:               # 2019년과 2020년의 비교 그래프이다.
plt.rc('font', family='Malgun Gothic')
fig = plt.figure(figsize=(10,6))
x = np.array(['2019', '2020'])
y = np.array([106.65,94.94])
plt.subplot(1, 2, 1)
plt.ylim([90,110])
plt.title('날씨가 있는 날')
plt.bar(x,y,width = 0.7)

#plot 2:
x = np.array(['2019', '2020'])
y = np.array([109.19,99.37])
plt.subplot(1, 2, 2) 
plt.ylim([90,110])
plt.title('날씨가 없는 날')
plt.bar(x,y,width = 0.7,color = 'red')

plt.show()
```
