```python
import matplotlib.pyplot as plt
import numpy as np
#plot 1:

plt.rc('font', family='Malgun Gothic')

fig = plt.figure(figsize=(10, 6))


x = np.array(['날씨가 있는 날', '날씨가 없는 날'])
y = np.array([106.64,109.25])
plt.subplot(1, 2, 1)
plt.ylim([80,115])
plt.title('2019년')
plt.bar(x,y)



#plot 2:





x = np.array(['날씨가 있는 날', '날씨가 없는 날'])
y = np.array([94.94,99.37])
plt.subplot(1, 2, 2) 
plt.ylim([80,115])
plt.title('2020년')
plt.bar(x,y)





plt.show()
```

```python
#plot 1:
plt.rc('font', family='Malgun Gothic')
fig = plt.figure(figsize=(10,6))
x = np.array(['2019', '2020'])
y = np.array([106.64,94.94])
plt.subplot(1, 2, 1)
plt.ylim([90,110])
plt.title('날씨가 있는 날')
plt.bar(x,y)

#plot 2:
x = np.array(['2019', '2020'])
y = np.array([109.25,99.37])
plt.subplot(1, 2, 2) 
plt.ylim([90,110])
plt.title('날씨가 없는 날')
plt.bar(x,y,color = 'red')

plt.show()
```
