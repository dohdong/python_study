```python
import numpy as np
import pandas as pd 
```

```python
fruit = pd.DataFrame(
    {'Num' : [123,456,789,1234,4353],
     'Fruit' : ['Apple','Banana','Cherry','Lemon','Peach']})
grade = pd.DataFrame(
    {'Num' : [123,456,743],
     'Grade':['a','b','c']})
price = pd.DataFrame(
    {'no':[123,456,2342],
     'Grade':[100,200,300]})
```

```python
fruit
```

```python
df = pd.concat([fruit,grade])
df   # 의미는 없음.
```

```python
df = pd.merge(fruit,grade,on = 'Num')
df
```

```python
df = pd.merge(fruit,grade,on = 'Num',how = 'left' )#left, right, outer 등이 있다.
df
```

```python
df = pd.merge(fruit, price, left_on = 'Num', right_on = 'no', how = 'outer')
df
```
